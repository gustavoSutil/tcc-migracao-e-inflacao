# Descrição

Nesta etapa, estou carregando os dados do Censo 2022 e as estimativas populacionais do IBGE. Em seguida, faço um ajuste nos números de população municipal com base na diferença entre os valores informados pelo censo e as estimativas anteriores do IBGE.

O objetivo é corrigir a população anual por município usando o censo como base confiável, garantindo que as projeções fiquem mais alinhadas com os dados oficiais mais recentes.

```python
from src.pre_process.load_data import load_censo_2022, diagnostico_csv, load_ibge
from src.pre_process.preprocess_pop import (
    construir_populacao_corrigida
)
diagnostico_csv('../data/raw/dados_censo_2022.csv', encoding='utf-8')

censo_2022 = load_censo_2022('../data/raw/dados_censo_2022.csv')
pop_ibge =  load_ibge('../data/raw/dataset_etimativa_pop_ibge_2010_2023.csv')
```

```python
print(pop_ibge.head())
print(censo_2022.head())
df_pop_corr = construir_populacao_corrigida(
    df_est=pop_ibge,
    df_censo=censo_2022
)

print(df_pop_corr.head())

df_pop_corr.to_csv(
    '../data/processed/pop_municipal_corrigido_anual_ibge.csv',
    index=False
)
```
